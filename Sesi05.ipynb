## Struktur Data di Pandas

### Series

```python
revenues = pd.Series([5555, 7000, 1980])
print(revenues, end='\n\n')
print(type(revenues), end='\n\n')
print(revenues.dtypes)
```

Menampilkan indeks posisi

```python
print("index:", revenues.index)
```

Menampilkan nilai dalam Series

```python
revenues.values
```

Menampilkan jumlah baris menggunakan method shape

```python
revenues.shape
```

Memberikan nama indeks (label) untuk setiap baris

```python
city_revenues = pd.Series(
    [4200, 8000, 6500],
    index=["Amsterdam", "Toronto", "Tokyo"],
    name="income"
)
print(city_revenues)
```

Membuat Series dengan indeks label dari dictionary

```python
city_employee_count = pd.Series(
    {"Amsterdam": 5, "Tokyo": 8},
    name="count"
)
print(city_employee_count)
```

### DataFrame

```python
city_data = pd.DataFrame({
    "revenue": city_revenues,
    "employee_count": city_employee_count
})
city_data
```

## Query DataFrame

Menampilkan data kolom sesuai kondisi tertentu

```python
df_gsw = df_summary[df_summary["team_id"] == "GSW"]
df_gsw
```

```python
df_gsw_win = df_summary[
    (df_summary["team_id"] == "GSW") & (df_summary.pts > df_summary.opp_pts)
]
df_gsw_win
```

```python
df_gsw_win["opp_id"].value_counts()
```

Mengurutkan baris (sorting)

```python
df_gsw_win.sort_values("pts", ascending=False)
```

```python
df_gsw_win.sort_values(["pts", "opp_pts"], ascending=[False, True])
```

## Grouping

```python
df.groupby("fran_id")[["pts", "opp_pts"]].sum()
```

```python
df_la_base_team = df.loc[
    df["team_id"].str.startswith("LA")
]
df_la_base_team
```

```python
df_la_base_team[
    df_la_base_team.year_id >= 2000
].groupby(["fran_id", "game_result"])["game_result"].count()
```

```python
df_la_base_team[
    (df_la_base_team.game_result == "L")
    & (df_la_base_team.fran_id == "Clippers")
    & (df_la_base_team.year_id >= 2000)
].shape[0]
```

```python
df_la_base_team[
    (df_la_base_team.game_result == "W")
    & (df_la_base_team.fran_id == "Clippers")
    & (df_la_base_team.year_id >= 2000)
].shape[0]
```
