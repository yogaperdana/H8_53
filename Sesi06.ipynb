## Unexpected Missing Values

Column `OWN_OCCUPIED`

```python
df["OWN_OCCUPIED"]
```

```python
df["OWN_OCCUPIED"].isna()
```

```python
df["OWN_OCCUPIED"].unique()
```

```python
df["OWN_OCCUPIED"].value_counts()
```

```python
df.loc[5, "OWN_OCCUPIED"]
```

```python
for idx, row in enumerate(df["OWN_OCCUPIED"]):
    if not isinstance(row, str):
        continue
    try:
        int(row)
        df.loc[idx, "OWN_OCCUPIED"] = np.nan
    except:
        pass

df["OWN_OCCUPIED"]
```

Column `NUM_BATH`

```python
df["NUM_BATH"]
```

```python
for idx, row in enumerate(df["NUM_BATH"]):
    if not isinstance(row, str):
        continue
    try:
        df.loc[idx, "NUM_BATH"] = int(float(row))
    except:
        df.loc[idx, "NUM_BATH"] = np.nan

df["NUM_BATH"]
```

```python
df
```

```python
df["ST_NUM"]
```

Column `OWN_OCCUPIED`:

```python
df["OWN_OCCUPIED"]
```

```python
df["OWN_OCCUPIED"].fillna(df["OWN_OCCUPIED"].mode().values[0], inplace=True)
df["OWN_OCCUPIED"]
```

Column `NUM_BEDROOMS`:

```python
df["NUM_BEDROOMS"]
```

```python
df["NUM_BEDROOMS"].fillna(df["NUM_BEDROOMS"].median(), inplace=True)
df["NUM_BEDROOMS"]
```

Column `NUM_BATH`:

```python
df["NUM_BATH"]
```

```python
int(df["NUM_BATH"].astype("float").mean())
```

```python
df["NUM_BATH"].fillna(int(df["NUM_BATH"].astype("float").mean()), inplace=True)
df["NUM_BATH"]
```

Column `SQ_FT`:

```python
df["SQ_FT"]
```

```python
df["SQ_FT"].fillna(df["SQ_FT"].mean(), inplace=True)
df["SQ_FT"]
```

## Time Series

### Basic Time Series Manipulation

Import library:

```python
from datetime import datetime
```

Membuat rentang tanggal dengan frekuensi harian:

```python
pd.date_range("2021/01/01", "2021/06/01", freq="D")
```

Membuat rentang tanggal dengan frekuensi per jam:

```python
daterange_hour = pd.date_range("2021/01/01", "2021/06/01", freq="H")
daterange_hour
```

Membuat dataframe dengan timestamp:

```python
df_time = pd.DataFrame(daterange_hour, columns=["date"])
df_time
```

Membuat kolom `data` dengan data angka acak:

```python
df_time["data"] = np.random.randint(100, size=df_time.shape[0])
df_time
```

Menampilkan tipe datanya:

```python
df_time.info()
```

Konversikan indeks data frame menjadi datetime index:

```python
df_time["datetime"] = pd.to_datetime(df_time["date"])
df_time.set_index("datetime", inplace=True)
df_time.drop("date", axis=1, inplace=True)
df_time
```

```python
#df_time.index.week
pd.Int64Index(df_time.index.isocalendar().week)
```

```python
df_time.index.day_of_year
```

```python
df_time.index.day_name()
```

```python
df_time.index.quarter
```

```python
df_time[(df_time.index.hour <= 12) & (df_time.index.hour >= 6)]
```

```python
df_time[df_time.index.is_month_end]
```

```python
df_time.loc["2021-01-01 01":"2021-01-01 12"]
```

```python
df_time.resample("W").sum()
```

```python
df_time.resample("2W").sum()
```

```python
daily_average = df_time.resample("D").mean()
biweekly_average = df_time.resample("2W").mean()
monthly_total = df_time.resample("M").sum()

display(daily_average.head())
display(biweekly_average.head())
display(monthly_total.head())
```

```python
daily_average.rolling(5).mean()
```

```python
daily_average.rolling("5D").mean()
```

```python
daily_average.rolling(5).mean().fillna(method="backfill")
```
